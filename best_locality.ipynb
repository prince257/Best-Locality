```python
# function that extracts the category of the venue.

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# clean the json and structure it to a dataframe
```

```python
venues = results['response']['groups'][0]['items']

nearby_venues = json_normalize(venues) # flatten JSON

# print(nearby_venues)

filtered_columns = ['venue.name' , 'venue.categories' , 'venue.location.lat' , 'venue.location.lng']
nearby_venues = nearby_venues.loc[: ,filtered_columns]

# nearby_venues.head()
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
# nearby_venues.head()
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
# function to get all nearby venues of each Neighborhood in Faridabad and making a new dataframe called faridabad_venues.
```

```python
def getNearbyVenues(names , latitude , longitude,radius = 1000):
    
    venues_list = []
    for name ,lat,lng in zip(names ,latitude ,longitude):
        print(name)
        
        url ='https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results
        ])
        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list] )
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)



        
```

```python
faridabad_venues = getNearbyVenues(names=faridabad_df['Neighborhood'],
                                   latitude=faridabad_df['latitude'],
                                   longitude=faridabad_df['longitude'],
                               radius = 1000
)
```

### Analyzing each Venue in all Neighborhoods by making dummies of "Venue Category" columns.

## Now, lets get the top 5 venues of each neighborhood in Faridabad.


```python
num_top_venues = 5 # number of top venues you want to find

for hood in faridabad_onehot_group["Neighborhood"]:
    print(f"-----{hood}-----")   
    temp = faridabad_onehot_group[faridabad_onehot_group["Neighborhood"]==hood].T.reset_index()
    temp.columns = ['Venue' ,'freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq' :2})
    print(temp.sort_values('freq' ,ascending = False).reset_index(drop = True).head(num_top_venues))
    print('/n')
    
```

## Now, find the 10 most common Venues in each Neighborhood.

```python
def return_most_common_venues(row,num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# creating columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
        
Neighborhood_venue_sorted = pd.DataFrame(columns = columns)
Neighborhood_venue_sorted['Neighborhood'] = faridabad_onehot_group['Neighborhood']

for i in np.arange(faridabad_onehot_group.shape[0]):
    Neighborhood_venue_sorted.iloc[i,1:] = return_most_common_venues(faridabad_onehot_group.iloc[i, 1:], num_top_venues)

    
Neighborhood_venue_sorted.head()                                 
```

```python
Neighborhood_venue_sorted.shape
```

### To get the idea of Neighborhoods in Faridabad i.e which Neighborhoods belong to each other on the basis of Nearby Venues we will use Kmeans Clustering Algorithm.

## Making Clusters containing Neighborhhods 

```python
number_of_clusters = 6 # number of total clusters we want.

faridabad_group_clustering = faridabad_onehot_group.drop('Neighborhood' , axis =1)

kmeans = KMeans(n_clusters = number_of_clusters ,random_state = 0).fit(faridabad_group_clustering)

kmeans.labels_
```

### Lets create a new dataframe which contain the cluster label of neighborhood with its 10 most common venues

```python
Neighborhood_venue_sorted.insert(0 , 'Cluster Labels' , kmeans.labels_)

faridabad_merged = faridabad_df

faridabad_merged = faridabad_merged.join(Neighborhood_venue_sorted.set_index("Neighborhood") ,on ='Neighborhood')
```
